```python
import pandas as pd
import numpy as np
import lib

# Read data and limit to 2003 to 2015
df = pd.DataFrame.from_csv('data.csv')
beg_year = 2003
end_year = 2016

df = df.loc[(df['Year'] >= beg_year) & (df['Year'] < end_year)]

metrics = ['VORP_Per_Year', 'BPM_Per_Year', 'WS/48']

results = []
for i in metrics:
    results.append(lib.sorted_value(df, i))
for i in metrics:
    results.append(lib.pick_value(df, i, beg_year, end_year))
results2 = pd.concat(results, axis=1)
results2 = results2.rank(ascending=False)
results2['Avg_Rank'] = results2.mean(axis=1)
results2['Final_Rank'] = results2['Avg_Rank'].rank().astype(int)
results2.sort_values(by='Final_Rank').style.background_gradient(cmap='RdYlGn_r', high=.2, low=.6)
```
