```python
# Simple Monte Carlo methods used

# Import libraries
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Roll 100-sided die
def random_number():
    roll = random.randint(1, 100)
    return roll

def Warriors_Win(num_sim, winning_percent):
    # Determine if it's a win or loss
    simulations = []
    for i in range(num_sim):
        Win_Loss = []
        for j in range(82):
            result = random_number()
            if result <= winning_percent:
                Win_Loss.append('Win')
            else:
                Win_Loss.append('Loss')
        simulations.append(Win_Loss)
    # Determiine if its a consecutive loss seasons    
    consecutive_loss_seasons = []
    z = 0
    while z < num_sim:
        # If any of the 82 games features consecutive losses, break the loop and append a 0 for the season
        for i in range(81):
            if (simulations[z][i] == 'Loss') & (simulations[z][i+1] == 'Loss'):
                consecutive_loss_seasons.append(0)
                break
        else:
            # If it doesn't, append a 1
            consecutive_loss_seasons.append(1)
        z += 1
    # Calculate running percentage of consecutive loss seasons
    percentages = []
    count = 0
    for i in range(len(consecutive_loss_seasons)):
        if consecutive_loss_seasons[i] == 1:
            count += 1
        percentages.append(float(count)/(i+1)*100)
    # Compute final result, standard deviation
    final_percent = percentages[-1]
    std = np.std(percentages)
    return final_percent, std, percentages

def generate_summary(num_sim, winning_percent):
    final_percent, std, percentages = Warriors_Win(num_sim, winning_percent)
    print "Warriors go without consecutive losses in ~{}% +/- {} of simulated seasons".format(final_percent, std)
    # Plot figure
    fig = plt.figure(figsize=(16, 8))
    fig.suptitle('Simulation Summary', fontsize=16)
    plt.xlabel('Number of Simulations', fontsize=16)
    plt.ylabel('Non-Consecutive Loss Seasons (%)', fontsize=16)
    plt.grid()
    plt.plot(percentages)
    return fig.show()

# Time it
import time
t0 = time.time()

# Run the simulation, 100,000 simulations takes ~15 seconds
generate_summary(num_sim=100000, winning_percent=80)

t1 = time.time()
total = t1-t0
print total, 'seconds to run'
```

```python
# Run this process for winning percentages 80-100
def iterator(num_sim):
    iterations = []
    for i in range(80, 101):
        final_percent, std, percentages = Warriors_Win(num_sim, i)
        iterations.append([i, final_percent])
    # Plot figure    
    fig, ax = plt.subplots(figsize=(20,10))
    ax.set_xticks(np.arange(80, 100))
    ax.set_yticks(np.arange(0, 100, 5))
    plt.suptitle('Determining Non-Consecutive Loss Season % for Winning Percentages 80 to 100', fontsize=16)
    plt.xlabel('Winning Percentage', fontsize=16)
    plt.ylabel('Non-Consecutive Loss Seasons (%)', fontsize=16)
    plt.grid()
    # X is possible range, Y is the final percent calculated
    plt.plot([item[0] for item in iterations], [item[1] for item in iterations])
    # Plot 50% line
    plt.plot([80, 100], [50, 50], 'k-', lw=2)
    return fig.show()

# Time and run: 10,000 simulations takes ~30 seconds
t0 = time.time()
iterator(num_sim=10000)

t1 = time.time()
total = t1-t0
print total, 'seconds to run'
```

```python
# Import libraries
from datetime import datetime
import pandas as pd
import numpy as np
from collections import defaultdict, OrderedDict, Counter

# Read data
division = pd.read_excel('Analytics_Attachment.xlsx', 0)
df = pd.read_excel('Analytics_Attachment.xlsx', 1)
eliminated = pd.read_excel('Analytics_Attachment.xlsx', 2)

# Turn Home/Away Winners into Team Winning
df_Home = df.loc[df['Winner'] == 'Home']
df_Away = df.loc[df['Winner'] == 'Away']
df_Home['Winning Team'] = df['Home Team']
df_Home['Losing Team'] = df['Away Team']
df_Away['Winning Team'] = df['Away Team']
df_Away['Losing Team'] = df['Home Team']
df = pd.concat([df_Home, df_Away]).sort_index()

# Zip Into Date/Team df
dates = set(df['Date'])
teams = set(df['Home Team'])
zipped = [(x,y) for x in dates for y in teams]
# Merge in division/conferences
df2 = pd.DataFrame(zipped, columns=['Date', 'Team'])
division = division.rename(columns={'Team_Name': 'Team'})
df2 = pd.merge(df2, division, on='Team')

# Count Wins of winning team
df_win = df[['Date', 'Winning Team']]
df_win['Wins'] = df_win.groupby('Winning Team').cumcount() + 1
df_win.rename(columns={'Winning Team':'Team'}, inplace=True)
df_wins = pd.merge(df2, df_win, on=['Date', 'Team'])

# Count losses of losing team
df_loss = df[['Date', 'Losing Team']]
df_loss['Losses'] = df_loss.groupby('Losing Team').cumcount() + 1
df_loss.rename(columns={'Losing Team':'Team'}, inplace=True)
df_losses = pd.merge(df2, df_loss, on=['Date', 'Team'])

# Combine wins/losses into date/team df
df3 = pd.concat([df_wins, df_losses])
df3 = pd.concat([df2, df3]).groupby(['Date','Team']).first().reset_index().sort_values(['Team','Date']).reset_index(drop=True)

# Fill in blanks for each team
team = df3['Team']
df3 = df3.groupby('Team').transform(lambda x: x.fillna(method='ffill')).fillna(0)
df3['Team'] = team

# Add relevant features
df3['Games Played'] = df3['Wins'] + df3['Losses']
df3['Games Remaining'] = 82 - df3['Games Played']
df3['Possible Wins'] = df3['Wins'] + df3['Games Remaining']

# Create Win, Loss, Future dataframes
df_winloss = df[['Date', 'Winning Team', 'Losing Team']]
df_winloss2 = df_winloss.rename(columns={'Winning Team':'Losses Against', 'Losing Team':'Team'})
df_future = df.rename(columns={'Winning Team':'Team', 'Losing Team':'Opponent'})
df_future2 = df.rename(columns={'Winning Team':'Opponent', 'Losing Team':'Team'})
df_future = pd.concat([df_future, df_future2]).sort_values('Date').reset_index(drop=True)
df_winloss.rename(columns={'Winning Team':'Team', 'Losing Team':'Wins Against'}, inplace=True)
df_winloss.set_index(['Team'], inplace=True)
df_winloss2.set_index(['Team'], inplace=True)
df_future.set_index(['Team'], inplace=True)

# Create lists of wins, losses, and future games
tot_wins = df_winloss['Wins Against'].groupby(df_winloss.index).agg(lambda x: list(list(x)))
tot_losses = df_winloss2['Losses Against'].groupby(df_winloss2.index).agg(lambda x: list(list(x)))
future_games = df_future['Opponent'].groupby(df_future.index).agg(lambda x: list(list(x)))

# Place in correct sorting order
df_winloss = df_winloss.reset_index().sort_values(['Team', 'Date'])
df_winloss2 = df_winloss2.reset_index().sort_values(['Team', 'Date'])
df_future = df_future.reset_index().sort_values(['Team', 'Date'])

# Create rolling Win_Loss record
def create_win_loss_record(tot_WL, df_WL, Wins_Losses_String):
    cum_list = []
    for i in range(len(tot_WL)):
        # Mini_lists are the rolling record
        mini_list = []
        for j in range(len(tot_WL[i])):
            mini_list.append(tot_WL[i][:j+1])
        cum_list.append(mini_list)
    # Flatten list
    cum_list = [item for sublist in cum_list for item in sublist]
    cum_count = []
    # Use counter to count appearances
    for i in cum_list:
        cum_count.append(Counter(i))
    df_WL['Cumulative {}'.format(Wins_Losses_String)] = cum_count
 
# Run for wins and losses
Wins = 'Wins'
Losses = 'Losses'
Opponents_Remaining = 'Opponents Remaining'   
create_win_loss_record(tot_wins, df_winloss, Wins)
create_win_loss_record(tot_losses, df_winloss2, Losses)

# Run for future games
cum_list = []
for i in range(len(future_games)):
    # Same thing, except in reverse
    mini_list = []
    for j in range(len(future_games[i])):
        mini_list.append(future_games[i][j+1:])
    cum_list.append(mini_list)
cum_list = [item for sublist in cum_list for item in sublist]
cum_count = []
for i in cum_list:
    cum_count.append(Counter(i))
    
df_future['Cumulative Opponents Remaining'] = cum_count

# Combine new dataframes with the date/team df
df4 = pd.merge(df3, df_winloss, on=['Date', 'Team'])
df5 = pd.merge(df3, df_winloss2, on=['Date', 'Team'])
df6 = pd.merge(df3, df_future, on =['Date', 'Team'])
df7 = pd.concat([df6, df5, df4, df3]).groupby(['Date', 'Team']).first().reset_index()

# Mark for filling gaps in date/time log
mark_ffill = df7[['Date', 'Team', 'Cumulative Wins', 'Cumulative Losses', 'Cumulative Opponents Remaining']]
team = mark_ffill['Team']
mark_ffill = mark_ffill.groupby('Team').transform(lambda x: x.fillna(method='ffill'))
mark_ffill['Team'] = team
df7 = pd.concat([mark_ffill, df7]).groupby(['Date', 'Team']).first().reset_index()

# Add feature
df7['Cumulative Potential Wins'] = df7['Cumulative Wins'] + df7['Cumulative Opponents Remaining']
df = df7.copy()

# Calculate 8th place stats in each conference
g = df.sort_values('Wins').groupby(['Date', 'Conference_id'])
eighth_place = g.nth(-8)['Wins']
eighth_place_team = g.nth(-8)['Team']
eighth_place_team_cum = g.nth(-8)['Cumulative Wins']
eighth_place_pot = g.nth(-8)['Possible Wins']
ninth_place_pot = g.nth(-9)['Possible Wins']
eighth_place = pd.DataFrame(eighth_place).reset_index()
eighth_place.rename(columns={'Wins':'Eighth_Place_Wins'}, inplace=True)
eighth_place_team = pd.DataFrame(eighth_place_team).reset_index()
eighth_place_team.rename(columns={'Team':'Eighth_Place_Team'}, inplace=True)
eighth_place_team_cum =  pd.DataFrame(eighth_place_team_cum).reset_index()
eighth_place_team_cum.rename(columns={'Cumulative Wins':'Eighth_Place_Team_Cum'}, 
                                      inplace=True)
eighth_place_pot =  pd.DataFrame(eighth_place_pot).reset_index()
ninth_place_pot = pd.DataFrame(ninth_place_pot).reset_index()
eighth_place_pot.rename(columns={'Possible Wins':'Eighth_Place_Pot'}, 
                                      inplace=True)
ninth_place_pot.rename(columns={'Possible Wins':'Ninth_Place_Pot'}, 
                                      inplace=True)

df = pd.merge(df, eighth_place, on=['Date', 'Conference_id'])
df = pd.merge(df, eighth_place_team, on=['Date', 'Conference_id'])
df = pd.merge(df, eighth_place_team_cum, on=['Date', 'Conference_id'])
df = pd.merge(df, eighth_place_pot, on=['Date', 'Conference_id'])
df = pd.merge(df, ninth_place_pot, on=['Date', 'Conference_id'])
df = df.sort_values('Date').reset_index(drop=True)

# Run through date/team df and go through (partial) checklist
# Not enough items added to distinguish Miami and Chicago
eliminated = []
clinched = []
for i in range(len(df)):
    # If Eighth Place current wins is greater than possible Team wins, Team is eliminated
    if df['Eighth_Place_Wins'][i] > df['Possible Wins'][i]:
        eliminated.append((df['Team'][i], df['Date'][i]))
    # Else if Ninth Place potential wins is less than current Team wins, Team clinches
    elif df['Ninth_Place_Pot'][i] < df['Wins'][i]:
        clinched.append((df['Team'][i], 'Playoffs'))
    # Else if Eighth Place current wins is equal team possible wins, 
    elif df['Eighth_Place_Wins'][i] == df['Possible Wins'][i]:
        # If Team Potential wins vs. Eighth Place Team is less than current Eighth Place current wins vs. Team,
        # Team is eliminated
        if df['Cumulative Potential Wins'][i][df['Eighth_Place_Team'][i]] < \
        df['Eighth_Place_Team_Cum'][i][df['Team'][i]]:
            eliminated.append((df['Team'][i], df['Date'][i]))

# Combine eliminated and clinched
eliminated = pd.DataFrame(eliminated, columns=['Team', 'Date'])
eliminated = eliminated.groupby(['Team']).first().reset_index()
clinched = pd.DataFrame(clinched, columns=['Team', 'Date'])
clinched = clinched.groupby(['Team']).first().reset_index()
total = pd.concat([eliminated, clinched]).sort_values('Team').reset_index(drop=True)

# Change date from DT format to Sample format
total['Date'] = total['Date'].astype(str)
dates = []
for i in total['Date']:
    i = i.split(' ')[0]
    if '-' in i:
        year = i.split('-')[0]
        month = i.split('-')[1]
        day = i.split('-')[2]
        dates.append(''.join(month + '/' + day + '/' + year))
    else:
        dates.append(i)
total['Date'] = dates
total.rename(columns={'Date': 'Date Eliminated'}, inplace=True)

print total

# Output to CSV
total.to_csv('Q2.csv')
```
